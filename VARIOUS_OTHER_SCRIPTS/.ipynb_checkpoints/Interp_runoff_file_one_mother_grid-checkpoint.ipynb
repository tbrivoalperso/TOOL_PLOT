```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import shutil
from shutil import copyfile
```

```python
def check_mask_nearby(i,j,mask,ndist):
    # i = index along i of point to test
    # j = index along j of point to test
    # mask = 2D mask 
    # ndist = distance from center to check
    if (i > (ndist-1)):
        TEST_i = np.nansum(mask[i-ndist:i+ndist+1,j])
    else:
        TEST_i = np.nansum(mask[i:i+ndist+1,j])
    if (j > (ndist-1)):
        TEST_j = np.nansum(mask[i,j-ndist:j+ndist+1])
    else:
        TEST_j = np.nansum(mask[i,j:j+ndist+1])
    return TEST_i, TEST_j    
```

```python
def find_nearest_sea_point(i,j,mask,ndist,rnf_data):
        arr= np.array([rnf_data[i-ndist,j] + mask[i-ndist,j], \
                    rnf_data[i+ndist,j]    + mask[i+ndist,j],    \
                    rnf_data[i,j-ndist]   + mask[i,j-ndist],   \
                    rnf_data[i,j+ndist]   + mask[i,j+ndist]])
        ind=np.array(np.argwhere(arr==0))
        if len(ind)==1:
            if ind == 0:
                return(i-ndist,j)
            if ind == 1:
                return(i+ndist,j)  
            if ind == 2:
                return(i,j-ndist)      
            if ind == 3:
                return(i,j+ndist)
            return ind
        else:
#             print("problem",len(ind) )
            out=[]
            if arr[0] == 0:
                out=np.append(out,[i-ndist,j])
            if arr[1] == 0:
                out=np.append(out,[i+ndist,j])  
            if arr[2] == 0:
                out=np.append(out,[i,j-ndist])      
            if arr[3] == 0:
                out=np.append(out,[i,j+ndist])
            return out
```

```python
def get_npoints(i,j,mask,ndist,rnf_data):
        # return the number of sea points at ndist distance
        arr= np.array([rnf_data[i-ndist,j] + mask[i-ndist,j], \
                    rnf_data[i+ndist,j]    + mask[i+ndist,j],    \
                    rnf_data[i,j-ndist]   + mask[i,j-ndist],   \
                    rnf_data[i,j+ndist]   + mask[i,j+ndist]])
        ind=np.array(np.argwhere(arr==0))
        return len(ind)
```

```python
def find_nearest_coast_point(i,j,mask,rnf_data,nseuil):
    out=[]
    n=1
    var=0
    while (var==0) and (n<nseuil):
        if mask[i+n,j]==1:
            if rnf_data[i+n-1,j] == 0.:
                out=np.append(out,[i+n-1,j])
        if mask[i-n,j]==1:
            if rnf_data[i-n+1,j] == 0.:
                out=np.append(out,[i-n+1,j])
        if mask[i,j+n]==1:
            if rnf_data[i,j+n-1] == 0.:
                out=np.append(out,[i,j+n-1])           
        if mask[i,j-n]==1:
            if rnf_data[i,j-n+1] == 0.:
                out=np.append(out,[i,j-n+1])           
        var=len(out)
        n=n+1

            
    return(out) # closest coastal point
        
```

```python
def move_runoffs(rnf_2D,mask_inp,lverb):
    nseuil=10
    TEST = np.array(np.where(rnf_2D.values>0))
    rnf_2D_new = rnf_2D
    mask_data=mask_inp
    mask_data=np.where(mask_data==1,0,1) # invert mask
    sea_point_not_found=0.
    ndist=1
    for n in range(len(TEST[0,:])):
    #     print(rnf_2D[TEST[0,n],TEST[1,n]].values)
        TEST_i, TEST_j = check_mask_nearby(TEST[0,n],TEST[1,n],mask_data,ndist)
        TEST_ij = TEST_i + TEST_j
        if(TEST_ij > 0): # at least one earth point nearby
            if (mask_data[TEST[0,n],TEST[1,n]]==1):
                if lverb: print("==> j, i = ", TEST[0,n],TEST[1,n], "is on mask point with", TEST_i -1 \
                      ,"earth points along i nearby and", TEST_j - 1,\
                     "points along j")
                if TEST_ij == 6: # Surrounded by earth
                    if (ndist > 5): 
                        print("WARNING !!! : closest sea point is over ",nseuil," point perimeter at i, j = ", TEST[0,n],TEST[1,n])

                    while sea_point_not_found == 0:
                        TEST_i, TEST_j = check_mask_nearby(TEST[0,n],TEST[1,n],mask_data,ndist)
                        TEST_ij = TEST_i + TEST_j
                        if TEST_ij == ((1+ndist*2)*2):  
                            if lverb: print(TEST[0,n],TEST[1,n],"surrounded by earth points, trying ndist=+1;", TEST_i, TEST_j,(1+ndist*2)*2)
                            az=1
                        else: # Not Surrounded by earth
                            sea_point_not_found=1.                        
                            ind = find_nearest_sea_point(TEST[0,n],TEST[1,n],mask_data,ndist,rnf_2D.values)
                            npoints = get_npoints(TEST[0,n],TEST[1,n],mask_data,ndist,rnf_2D.values)
                        ndist=ndist+1
                else: # Not on mask point
                    ind = find_nearest_sea_point(TEST[0,n],TEST[1,n],mask_data,ndist,rnf_2D.values)
                    npoints = get_npoints(TEST[0,n],TEST[1,n],mask_data,ndist,rnf_2D.values)               
                # Now replace data
                if npoints == 0: # No available points nearby without an already existing runoff value => avoid
                    rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.  
                    if lverb: print("Surrounded by valid runoff value, next ")
                if npoints == 1: 
                    rnf_2D_new[int(ind[0]),int(ind[1])] = rnf_2D_new[TEST[0,n],TEST[1,n]]
                    rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.   
                    if lverb: print("moving point j,i = ", TEST[0,n],TEST[1,n], "at point j,i = ", ind)
                if npoints == 2: 
                    rnf_2D_new[int(ind[0]),int(ind[1])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 2 
                    rnf_2D_new[int(ind[2]),int(ind[3])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 2 
                    if lverb: print("dispatch point j,i = ", TEST[0,n],TEST[1,n], "at 2 points j1,i1,j2,i2 = ", ind)
                    rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.               
                if npoints == 3: 
                    rnf_2D_new[int(ind[0]),int(ind[1])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 3 
                    rnf_2D_new[int(ind[2]),int(ind[3])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 3 
                    rnf_2D_new[int(ind[3]),int(ind[4])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 3 
                    if lverb: print("moving point j,i = ", TEST[0,n],TEST[1,n], "at 3 points j1,i1,j2,i2,j3,i3 = ", ind)

                    rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.                    
                if lverb: print("..........")


        else: # no earth point nearby

            if lverb: print(TEST[0,n],TEST[1,n], "is in the middle of the sea", TEST_i, TEST_j,mask_data[TEST[0,n]+1,TEST[1,n]])
            ind = find_nearest_coast_point(TEST[0,n],TEST[1,n],mask_data,rnf_2D.values,nseuil)
            npoints=len(ind)/2
            if npoints == 0: # No available points nearby without an already existing runoff value => avoid
                rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.  
                print("WARNING : point at i, j = ", TEST[0,n],TEST[1,n] ," is way too far from coast => ignore")
            if npoints == 1: 
                rnf_2D_new[int(ind[0]),int(ind[1])] = rnf_2D_new[TEST[0,n],TEST[1,n]]
                rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.   
                if lverb: print("moving point j,i = ", TEST[0,n],TEST[1,n], "at point j,i = ", ind)
            if npoints == 2: 
                rnf_2D_new[int(ind[0]),int(ind[1])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 2 
                rnf_2D_new[int(ind[2]),int(ind[3])] = rnf_2D_new[TEST[0,n],TEST[1,n]] / 2 
                if lverb: print("dispatch point j,i = ", TEST[0,n],TEST[1,n], "at 2 points j1,i1,j2,i2 = ", ind)
                rnf_2D_new[TEST[0,n],TEST[1,n]] = 0.               
            if lverb: print("..........")

        sea_point_not_found=0.
        ndist=1
        # Next point
    return rnf_2D_new
```

```python
year=2017

rnf_inp_file='/data/rd_exchange/tbrivoal/RUNOFFS_eNEATL36/runoff_eNEATL36_with_BDY_y'+str(year)+'.nc'
inp_domain_cfg_file='/data/rd_exchange/tbrivoal/RUNOFFS_eNEATL36/domain_cfg.nc'
out_domain_cfg_file='/data/rd_exchange/tbrivoal/RUNOFFS_eNEATL36/NEW_DOMAIN_CFG/domain_cfg.nc'
out_domain_cfg_zoom_file='/data/rd_exchange/tbrivoal/RUNOFFS_eNEATL36/NEW_DOMAIN_CFG/1_domain_cfg.nc'
mask_file='/data/rd_exchange/tbrivoal/RUNOFFS_eNEATL36/mesh_mask.nc'
output_folder="/data/rd_exchange/tbrivoal/RUNOFFS_eNEATL36/"
```

```python
# Open variables
mask_inp =  inp_domain_cfg.top_level.squeeze()
mask_inp =  out_domain_cfg.top_level.squeeze()

# mask_inp = maskf.tmask[0,0,:,:].squeeze()
rnf_3D=inp_rnf_2D.orca_costal[:,:,:]
rnf_3D_new=rnf_3D
```

```python
year=2017 # year of runoff file

# Date variables ----------------------------------------------------------------------------------------
dstart_year= str(year) + "-01-01" # 1st day of the year
dend_year= str(year) + "-12-31" # last day of the year

dstart_clim_monthly = str(year) + "-01-15" # for 12 month clim
dend_clim_monthly = str(year) + "-12-15" # for 12 month clim
dstart_minus_1_month = str(year-1) + "-12-01" # used for conversion from monthly to daily clim data

############################ Save in a Netcdf ###################################
time_counter_daily = pd.date_range(dstart_year,dend_year, freq="D") # time dimension

# time_counter_daily_float = pd.to_numeric(time_counter_daily, downcast='float')
#time_counter_daily.values.astype("float")
# Create xarray dataarray
rnf_3D_daily_new_da = xr.DataArray(
    data=rnf_3D_new,
    dims=["time_counter", "y", "x"],
    coords=dict(time_counter=time_counter_daily.values),
    attrs=rnf_3D.attrs,
    name=rnf_3D.name,
     )



# Create dataset
ds = xr.merge([rnf_3D_daily_new_da, inp_rnf_2D.nav_lat, inp_rnf_2D.nav_lon])
ds.attrs=inp_rnf_2D.attrs
# Save to netcdf
outfile = output_folder + "runoff_eNEATL36_with_BDY_y" + str(year) + '_extrap_on_newdomaincfg.nc'  
ds.to_netcdf(outfile, mode="w")
```
