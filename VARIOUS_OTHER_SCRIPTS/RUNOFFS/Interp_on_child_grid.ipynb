```python
def find_nearest_coast_point(i,j,mask,nseuil):
    out=[]
    n=1
    var=0
    while (var==0) and (n<nseuil):
        if mask[i+n,j]==1:
            out=np.append(out,[i+n-1,j])
        if mask[i-n,j]==1:
            out=np.append(out,[i-n+1,j])
        if mask[i,j+n]==1:
            out=np.append(out,[i,j+n-1])
        if mask[i,j-n]==1:
            out=np.append(out,[i,j-n+1])  

        n=n+1
        var=len(out)
    return(out) # closest coastal point
        
```

```python
def count_land_points(i_mother,j_mother,i_zoom,j_zoom,mask_zoom,Nfactor):
    if (Nfactor % 2) == 0:
        print("STOP - REFINEMENT FACTOR IS PAIR - CASE NOT CODED YET")
        exit()
    else:
        ndist= int((Nfactor -1)/2) # (Number of points in mesh along i or j direction / 2) - the central point
        
    sum_mask=np.nansum(mask_zoom[i_zoom-ndist:i_zoom+ndist+1,j_zoom-ndist:j_zoom+ndist+1])

    return sum_mask
```

```python
def place_and_expand_runoffs_on_zoom(rnf_2D_mother_on_zoom,rnf_2D_zoom,rnf_point_mother,coordcheck=False,\
                                     nav_lat_zoom=None,nav_lon_zoom=None,nav_lat_mother=None,\
                                     nav_lon_mother=None):
    i_mother = rnf_point_mother[0,:]
    j_mother = rnf_point_mother[1,:]
    i_zoom = Nfactor * (i_mother) + 2 # +2 in order to be at f point and avoid rim point (+1 +1)
    j_zoom = Nfactor * (j_mother) + 2
    ndist= int((Nfactor -1)/2)
    nseuil=30
    number=0    
    mask_zoom_inv=np.where(mask_zoom==1,0,1) # invert mask
    for n in range(len(i_mother)):
    ############### TO TEST INDEXATION ---------------------------------------------------------------------
        if coordcheck:   print(nav_lat_mother[int(i_mother[n]),int(j_mother[n])].values,                 \
              nav_lat_zoom[int(i_zoom[n]),int(j_zoom[n])].values)
        if coordcheck:      print(nav_lon_mother[int(i_mother[n]),int(j_mother[n])].values,              \
              nav_lon_zoom[int(i_zoom[n]),int(j_zoom[n])].values)
    #------------------------------------------------------------------------------------------------------

        N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])                 \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor)
        if N_land_points == 9: 
            print("ERROR at, i,j (mother) = ", i_mother[n]+ymin,j_mother[n]+xmin,                        \
                  "and i, j (zoom) =",i_zoom[n],j_zoom[n] )
            print("This point is a sea point in mother grid but full of land points in zoom ") 
            print("Perhaps the input runoff file has been odly processed") 
        elif N_land_points == 0:
    #         print("point at, i,j (mother) = ", i_mother[n]+ymin,j_mother[n]+xmin,                      \
    #               "and i, j (zoom) =",i_zoom[n],j_zoom[n], 'is offshore' )
            ind = find_nearest_coast_point(i_zoom[n],j_zoom[n],mask_zoom_inv,nseuil)  
            if len(ind)==2:
                # Replace i_zoom and j_zoom by the nearest coast point and recount number of land points
                i_zoom[n]=ind[0] 
                j_zoom[n]=ind[1]
                N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])         \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor)
                mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                  \
                                                     j_zoom[n]-ndist:j_zoom[n]+ndist+1]
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=      \
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+      \
                         np.where(mask_zoom_around_point==0,                                             \
                         rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]\
                                  /(9-N_land_points),0.)            

            elif len(ind)==4:
                # Dispatch over 2 points
                # First point ---------------------------------------------- 
                i_zoom[n]=ind[0] 
                j_zoom[n]=ind[1]
                N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])           \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor) 
                mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                    \
                                                     j_zoom[n]-ndist:j_zoom[n]+ndist+1]
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=        \
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+        \
                         np.where(mask_zoom_around_point==0,                                               \
                         rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]                        \
                                  /((9-N_land_points)* 2),0.)         
                # Second point ---------------------------------------------- 

                i_zoom[n]=ind[2] 
                j_zoom[n]=ind[3]
                N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])           \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor)  
                mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                    \
                                                     j_zoom[n]-ndist:j_zoom[n]+ndist+1]
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=        \
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+        \
                         np.where(mask_zoom_around_point==0,                                               \
                         rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]                        \
                                  /((9-N_land_points)* 2),0.)          
            elif len(ind)==6:
                # Dispatch over 3 points
                # First point ---------------------------------------------- 
                i_zoom[n]=ind[0] 
                j_zoom[n]=ind[1]
                N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])           \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor) 
                mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                    \
                                                     j_zoom[n]-ndist:j_zoom[n]+ndist+1]
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=        \
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+        \
                         np.where(mask_zoom_around_point==0,                                               \
                         rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]                        \
                                  /((9-N_land_points)* 3 ),0.)      

    #             # Second point ---------------------------------------------- 

                i_zoom[n]=ind[2] 
                j_zoom[n]=ind[3]
                N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])          \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor) 
                mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                   \
                                                     j_zoom[n]-ndist:j_zoom[n]+ndist+1]
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=       \
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+       \
                         np.where(mask_zoom_around_point==0,                                              \
                         rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]                       \
                                  /((9-N_land_points)* 3),0.)            
    #             print("N=", N_land_points,rnf_2D_zoom[i_zoom[n],j_zoom[n]],i_zoom[n],j_zoom[n],rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])].values)

                # Second point ---------------------------------------------- 

                i_zoom[n]=ind[4] 
                j_zoom[n]=ind[5]
                N_land_points=count_land_points(int(i_mother[n]),int(j_mother[n]),int(i_zoom[n])          \
                                        ,int(j_zoom[n]),mask_zoom_inv,Nfactor)  
                mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                   \
                                                     j_zoom[n]-ndist:j_zoom[n]+ndist+1]
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=       \
                rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+       \
                         np.where(mask_zoom_around_point==0,                                              \
                         rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]                       \
                                  /((9-N_land_points)* 3),0.)   
            else: 
                print("ERROR - 4 points around, abort") 
                exit()

        else:
            mask_zoom_around_point=mask_zoom_inv[i_zoom[n]-ndist:i_zoom[n]+ndist+1,                    \
                                                 j_zoom[n]-ndist:j_zoom[n]+ndist+1]        
            rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]=        \
            rnf_2D_zoom[i_zoom[n]-ndist:i_zoom[n]+ndist+1,j_zoom[n]-ndist:j_zoom[n]+ndist+1]+        \
                     np.where(mask_zoom_around_point==0,                                               \
                     rnf_2D_mother_on_zoom[int(i_mother[n]),int(j_mother[n])]                        \
                              /(9-N_land_points),0.)            
    #     print(rnf_2D_zoom[i_zoom[n],j_zoom[n]])
    return rnf_2D_zoom
```

```python
year=2017 # year of runoff file

# Date variables ----------------------------------------------------------------------------------------
dstart_year= str(year) + "-01-01" # 1st day of the year
dend_year= str(year) + "-12-31" # last day of the year

dstart_clim_monthly = str(year) + "-01-15" # for 12 month clim
dend_clim_monthly = str(year) + "-12-15" # for 12 month clim
dstart_minus_1_month = str(year-1) + "-12-01" # used for conversion from monthly to daily clim data

############################ Save in a Netcdf ###################################
time_counter_daily = pd.date_range(dstart_year,dend_year, freq="D") # time dimension
print(time_counter_daily)
# time_counter_daily_float = pd.to_numeric(time_counter_daily, downcast='float')
#time_counter_daily.values.astype("float")
# Create xarray dataarray
rnf_2Dt_daily_zoom_da = xr.DataArray(
    data=rnf_2Dt_zoom,
    dims=["time_counter", "y", "x"],
    coords=dict(time_counter=time_counter_daily.values),
    attrs=rnf_2Dt_mother.attrs,
    name=rnf_2Dt_mother.name,
     )

nav_lat_da = xr.DataArray(
    data=nav_lat_zoom.values,
    dims=["y", "x"],
    attrs=nav_lat_mother.attrs,
    name=nav_lat_mother.name,
     )

nav_lon_da = xr.DataArray(
    data=nav_lon_zoom.values,
    dims=["y", "x"],
    attrs=nav_lon_mother.attrs,
    name=nav_lon_mother.name,
     )
# Create dataset
ds = xr.merge([rnf_2Dt_daily_zoom_da, nav_lat_da, nav_lon_da])
ds.attrs=inp_rnf_2Dt.attrs
# Save to netcdf
outfile = output_folder + "1_runoff_eNEATL36_with_BDY_y" + str(year) + '_extrap_on_newdomaincfg.nc'  
# outfile = output_folder + "runoff_eNEATL36_with_BDY_y" + str(year) + '.nc'  

ds.to_netcdf(outfile, mode="w")
```
